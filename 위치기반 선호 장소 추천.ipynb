## **2. 평균 구하고 군집별 최댓값 출력**

```python
avg_df = data.sort_values('Cluster')
avg_df
```

```python
# col 생략 없이 출력
pd.set_option('display.max_columns', None)
```

```python
avg = avg_df.groupby(['Cluster']) \
            .mean()
            # .max(axis=1)
avg
```

```python
# 각 그룹별로 가장 높은 평균과 해당 가게코드 찾기
max_means = avg.max(axis=1)
max_store_codes = avg.idxmax(axis=1)

# 각 그룹별 가장 높은 평균과 해당 가게코드 출력
for cluster, max_mean, max_store_code in zip(max_means.index, max_means, max_store_codes):
    print(f"그룹 {cluster}의 가장 높은 평균: {max_mean}, 가장 높은 평균을 가지는 가게코드: {max_store_code}")
```

## **4. 새 사용자 들어왔을 때, 최댓값의 가게 추천**

```python
# 새 사용자의 데이터 받기
new_user = pd.read_excel('/content/drive/MyDrive/위치/새 사용자.xlsx')
# new_user
new_user_co = new_user.drop('U_id', axis=1)  # 새 사용자의 유저 아이디 없앰.
new_user_co
```

```python
print(len(new_data))
print(len(new_user_co))
```

```python
# 기존 데이터와 새 사용자의 df 합침
result1 = pd.concat([new_data, new_user_co])
result1
```

```python
# kmeans 적용
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(result1)

# 클러스터링 결과를 DataFrame에 추가
result1['Cluster'] = clusters
result1
```

```python
new_users_clust = result1.iloc[-1,:]['Cluster']
new_users_clust  # Cluster 2

# new_user['Cluster'] = new_users_clust
# new_user
```

```python
new_user_avg_df = result1.sort_values('Cluster')
new_user_avg_df
```

```python
new_user_avg = new_user_avg_df.groupby(['Cluster']) \
                              .mean()
            # .max(axis=1)
new_user_avg
```

```python
# 각 그룹별로 가장 높은 평균과 해당 가게코드 찾기
new_user_max_means = new_user_avg.max(axis=1)

new_user_max_store_codes = []
for index, row in new_user_avg.iterrows():
    max_value = row.max()
    max_columns = row.index[row == max_value].tolist()
    new_user_max_store_codes.append(max_columns)

# 각 그룹별 가장 높은 평균과 해당 가게코드 출력
for cluster, new_user_max_mean, new_user_max_store_code in zip(new_user_max_means.index, new_user_max_means, new_user_max_store_codes):
    print(f"그룹 {cluster}의 가장 높은 평균: {new_user_max_mean}, 가장 높은 평균을 가지는 가게코드: {new_user_max_store_code}")
```
